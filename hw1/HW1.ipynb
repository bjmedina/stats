```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x):
    'Function that we want to minimize'
    return -2*torch.exp(-((x[0]-1.0)**2) - ((x[1]-1.0)**2)) - torch.exp(-((x[0]+2)**2) - ((x[1]+3)**2))
```

### b)

```python
# define list of initial x values
xs = [torch.tensor([0.0,0.0]), torch.tensor([-1.0,-1.0])]

for x0 in xs:
    x_0 = torch.autograd.Variable(x0, requires_grad=True)
    
    stored_x1 = []
    stored_x2 = []
    f_X       = []
    
    # gradient descent parameters
    step_sz    = 10e-12
    print_step = 500
    num_steps  = 8000

    for step in range(num_steps+1):
        # get the gradient (torch.autograd.grad returns 1-element tuple so need to grab element 0 of the output)
        x_grad = torch.autograd.grad(f(x_0), x_0)[0]

        x_0.data -= step_sz * x_grad

        if step % print_step == 0:
            print("Step: {} \t X: {} \t f(X): {}".format(step, x_0.data, f(x_0)))

        stored_x1.append(x_0[0].item())
        stored_x2.append(x_0[1].item())
        f_X.append(f(x_0))
        
    fig = plt.figure(figsize=(12,10))
    ax  = fig.gca(projection='3d')
    ax.set_title("Initial X: {}".format(x_0.detach()))
    ax.plot(stored_x1, stored_x2, zs=min(f_X).detach().item(), zdir='z', label='curve of X values')
    #print(stored_x1)
    #plt.plot(stored_x1, stored_x2)
    ax.plot(stored_x1, stored_x2, zs=f_X, zdir='z', label='f(x) values')
    ax.legend()

    ax.set_xlabel("X1")
    ax.set_ylabel("X2")
    ax.set_zlabel("f(X)")
    plt.show()
```

- What is a good step size for a reasonably fast convergence? 
    - $\eta=10^{-12}$ converges quickly and doesn't allow for exploding (or vanishing) gradients
- How many steps does convergence take?
    - Convergence takes about 4000 steps
