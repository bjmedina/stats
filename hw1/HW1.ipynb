```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    'Function that we want to minimize'
    
    return -2*torch.exp((x[0]-1)**2 + (x[1]-1)**2) - torch.exp((x[0]+2)**2 + (x[1]+3)**2)
```

### b)

```python
x_1 = torch.tensor([0.0,0.0])
x_2 = torch.tensor([-1.0,-1.0])

x1 = torch.autograd.Variable(x_1, requires_grad=True)
x2 = torch.autograd.Variable(x_2, requires_grad=True)

x1_1 = []
x2_1 = []
x1_2 = []
x2_2 = []

# gradient descent parameters
step_sz    = 1e-11
print_step = 100
num_steps  = 4000

for step in range(num_steps):
    # get the gradient (torch.autograd.grad returns 1-element tuple so need to grab element 0 of the output)
    x_grad1 = torch.autograd.grad(f(x1), x1)[0]
    x_grad2 = torch.autograd.grad(f(x2), x2)[0]

    x1.data -= step_sz * x_grad1
    x2.data -= step_sz * x_grad2

    if step % print_step == 0:
        print("F1: {} \t F2: {}".format(x1.data, x2.data))
        
    x1_1.append(x1.data[0])
    x1_2.append(x1.data[1])
    x2_1.append(x2.data[0])
    x2_2.append(x2.data[1])
```
