### a)

```python
%%R
## loading dataset
df_1 = read.csv("midterm_data_1.csv")

## It seems to me that the row column is unneccesary, so we'll remove that column
df_1$row = NULL

## Removing rows with NA
df_1 = na.omit(df_1)

print(str(df_1))
```

```python
%%R
## making feat.c and feat.g categorical variables
df_1$feat.c = as.factor(df_1$feat.c)
df_1$feat.g = as.factor(df_1$feat.g)

print(str(df_1))
```

```python
%%R 
## splitting dataset into 80/20
set.seed(005)

# 80/20 split for test-val split
smp_size = floor(0.80 * nrow(df_1))

# The test-val split
train_sample = sample(1:nrow(df_1), size=smp_size, replace=FALSE)

df_1_train = df_1[train_sample,]
df_1_test = df_1[-train_sample,]

print(nrow(df_1))
print(nrow(df_1_train))
print(nrow(df_1_test))
```

```python
%%R -w 9 -h 7 --units in -r 400
## pairwise plots between all columns
plot(df_1)
```

```python
%%R 
cat_col_names  = c("feat.c", "feat.g")
df_1_numerical = df_1[,-which(names(df_1) %in% c("feat.c", "feat.g"))]
print(str(df_1_numerical))
print(cor(df_1_numerical))
```

```python
%%R
lm1_inter_quad = lm(response ~ (feat.a + feat.b + feat.c + feat.d + feat.e + feat.f + feat.g + feat.h + feat.i)^2 
                    + I(feat.a^2) + I(feat.b^2) + I(feat.d^2) 
                    + I(feat.e^2) + I(feat.f^2) + I(feat.h^2) + I(feat.i^2),
                   data=df_1_train)
print(summary(lm1_inter_quad))
```

**A few statistical significant set of coefficients are of feat.a, feat.a and feat.b, feat.b and feat.d**

```python
%%R
lm1_reduced = lm(response ~ feat.a + (feat.a + feat.b)^2 + (feat.b + feat.d)^2, data=df_1_train)
print(summary(lm1_reduced))
```
